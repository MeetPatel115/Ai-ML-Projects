```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("president_polls_2016.csv")
```

```python
df[['state','grade']].value_counts()
```

```python
category_counts = df['state'].value_counts().reset_index(name='Count_Category')
value_counts = df['grade'].value_counts().reset_index(name='Count_Value')

# Display the counts
print(category_counts)
print(value_counts)
```

```python
combination_counts = df.groupby(['state','grade']).size().reset_index(name='Count')
```

```python
print(combination_counts)
```
