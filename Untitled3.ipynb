```python
df['native.country'].value_counts()
```

```python
threshold=500
# Identify main features and combine others
main_labels = [df['native.country'].value_counts().index if size >= threshold else 'Others' for label, size in zip(df['native.country'].value_counts().index, df['native.country'].value_counts().values)]
main_sizes = [df['native.country'].value_counts().values if size >= threshold else 0 for size in df['native.country'].value_counts().values]

# Calculate the size of the 'Others' category
others_size = sum(size for size in main_sizes if df['native.country'].value_counts().values < threshold)

# Add the 'Others' category to the main features
main_labels.append('Others')
main_sizes.append(others_size)

# Plot
plt.pie(main_sizes, labels=main_labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the plot
plt.show()
```
