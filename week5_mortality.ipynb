## how to Drop missing values

```python
data.dropna()
data
```

```python
data.head()
```

```python
# drop the row eith two or more missing values

print(data[(data['Year'].isnull()) & (data['DeathRate'].isnull())])

# other type
print(data[data.isnull().sum(axis=1)>=2])
```

```python
data[(data['Year'].isnull()) & (data['DeathRate'].isnull())].drop(inplace=False, axis=1)
```

```python
data.head(10)
```

```python
# drop the value with 2 by threshold tresh

data.dropna(thresh=2,inplace=False)
```

```python
data.drop
```

## How to impute the missing VAlues

```python
data['DeathRate'].mean()
```

```python
data['DeathRate'].fillna(value = data['DeathRate'].mean().round() , inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
## impute the missing values in year by Forward fill:

data.fillna(method= 'ffill',limit=2,inplace=True)
```
